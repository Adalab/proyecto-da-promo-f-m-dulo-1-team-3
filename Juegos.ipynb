### Bienvenid@ a la JUEGOS
Entra en la lanzadera de juegos y elige a qué juego quieres jugar:
- PIEDRA, PAPEL O TIJERA
- AHORCADO


```python
import random

class Juegos:
    def __init__(self):
        pass

    def Lanzadera_Juegos(self):
        jugar_ahora = input("Hola, ¿qué te apetece jugar hoy? (Piedra,papel o tijera/ Ahorcado)").upper()
                    
        if jugar_ahora == "PIEDRA,PAPEL O TIJERA":
            self.jugar_ppt()
        elif jugar_ahora == "AHORCADO":
            self.jugar_ahorcado()   
        elif jugar_ahora == "NO":
            print("👾 ¡Hasta pronto! 🤗")  

        # Ponemos un "return" sin llamar a una función corta el bucle:
        return   
        
    
    #Para jugar a PIEDRA PAPEL O TIJERA, ejecutamos la función:

    def jugar_ppt(self,puntos_jugador=0,puntos_pc=0):

        print('\n¡¡¡¡ Bienvenid@ al juego de piedra🪨, papel🗒️ o tijera✂️ !!!!\n', flush= True)
        nombre_jugador = input('¿Cómo quieres qué te llamemos?')
        print(f'Hola {nombre_jugador}. Comencemos a jugar. Vas a jugar contra el PC.\n Para ganarle tendrás que acumular tres puntos\n', flush= True)

        opciones= ['piedra', 'papel', 'tijera']
        puntos_jugador = 0
        puntos_pc = 0

        while puntos_jugador < 3 and puntos_pc < 3:

            jugador_usuario = input(f'Ingrese una opción entre las siguientes: {opciones}').lower()
            

            while jugador_usuario not in opciones: 
                print(f"La opción: {jugador_usuario} es una opcion equivocada. Ingrese una opción válida: {opciones}", flush= True)
                print ("___________________\n", flush= True)
                jugador_usuario = input(f'Ingrese una opción válida: {opciones}').lower() 

            jugador_pc = random.choice(opciones) 
            print (f"{nombre_jugador} eligió {jugador_usuario} y el PC {jugador_pc}", flush= True)
            
            if jugador_usuario == jugador_pc:
                print ("EMPATE", flush= True) 

            if (jugador_usuario == 'piedra' and jugador_pc == 'tijera') or (jugador_usuario == 'tijera' and jugador_pc == 'papel') or (jugador_usuario == 'papel' and jugador_pc == 'piedra'):
                print (f"GANA {nombre_jugador}", flush= True)
                puntos_jugador += 1

            if (jugador_usuario == 'piedra' and jugador_pc == 'papel') or (jugador_usuario == 'tijera' and jugador_pc == 'piedra') or (jugador_usuario == 'papel' and jugador_pc == 'tijera'):
                print ("GANA EL PC", flush= True)
                puntos_pc += 1

            print (f"Puntos {nombre_jugador}: {puntos_jugador} - Puntos PC: {puntos_pc}", flush= True)
            print ("________________________\n", flush= True)
            
        if puntos_jugador == 3 :
            print ("ENHORABUENA!! Has ganado la partida!!", flush= True)

        if puntos_pc == 3 :
            print ("Lo sentimos, has perdido la partida! Vuelve a intentarlo.", flush= True)

        
        automatizar = input("quieres volver a jugar?: Y/N").upper()
        while automatizar == "Y":
            print()
            self.jugar_ppt()
            automatizar = input("quieres volver a jugar?: Y/N").upper()

        if automatizar != "Y":
            print(flush = True)
            print("😊 BYE BYE 😊")

        return    
    
    #Para jugar a Ahorcado, ejecutamos las funciones: 

    def horca(self,vidas):
       
        print("   _______   ",flush = True)
        print(" |/       |  ",flush = True)
        if vidas <=7: 
            print(" |       (_)",flush= True)
        if vidas <=6: 
            print(" |        |",flush= True)
        else:
            print(" |           ",flush= True)
        if vidas <=5:
            print(" |       \|/",flush= True)
        else:
            print(" |           ",flush= True)
        if vidas <=4:
            print(" |        |",flush= True)
        else:
            print(" |           ",flush= True)  
        if vidas <=3:
            print(" |        |",flush= True)
        else:
            print(" |          ",flush= True)    
        if vidas ==2:
            print(" |       /  ",flush= True)      
        elif vidas <=1:
            print(" |       / \ ",flush= True)
        else:
            print(" |           ",flush= True)
        
        if vidas == 0:
            print(flush= True)
            print("GAME OVER! 🙃 ",flush= True)

    

    def jugar_ahorcado(self):
         
        horca = 0 
        opciones = ["melocoton", "tomate", "guayaba", "mandarina", "pomelo", "albaricoque", "frambuesa", "mangostino", "guanabana", "durian", "carambola", "nispero"]
        secreta = random.choice(opciones)

        secreta_ = list(secreta)                            
        longitud = len(secreta_)
        tablero_ = list ("_" * len(secreta_)) # Crear el tablero (ocultar palabra secreta)
        print(f"\n¡¡¡¡ Bienvenid@ al juego del ahorcado!!!!\n ARE U READY? \n Adivina la fruta. Tu tablero tiene {longitud} caracteres: {tablero_}",flush = True)
        print()
        print(" Si fallas, irá apareciendo un ahorcado.Tu misión es adivinar la palabra ¡sin ahorcarte! 🥵 ¡SUERTE!")

        respuesta = "".join(tablero_)  

        vidas = 8 

        

        while respuesta != secreta and vidas > 0:

            j2 = input ("Incluye una letra para adivinar la palabra secreta").lower() # Preguntamos al usuario una letra para adivinar la palabra secreta
                
            if j2 in secreta_:
                print (f"\nJUGADA:\nLa letra elegida es *** {j2} ***. Has acertado. Vuelve a insertar otra letra\n")
                print (f"Te quedan {vidas} vidas\n")
                for index, i in enumerate(secreta_):
                    if j2 == i:
                        tablero_[index] = i
                respuesta = "".join(tablero_)    #Lo indexamos al nivel del for para que me printe el tablero una vez finalize de iterar el "for"
                print (tablero_,flush = True)    
                            
            else:   # todos los otros casos donde j2 no está en secreta

                # Aquí, con cada fallo, se resta un intento
                vidas -= 1 
                            
                if vidas == 0:
                    print(f"😓 {vidas} vidas. ¡AHORCADO! 😵")
                    self.horca(vidas)

                elif vidas == 1:
                    print(f"😓 {vidas} vidas. Ten cuidado, la soga aprieta❗",flush= True)
                    self.horca(vidas)    
                    
                else:
                    #print ("Prueba otra letra",flush = True)
                    print(f"Prueba otra letra. Te quedan {vidas} vidas",flush= True) 
                    self.horca(vidas)  

        if respuesta == secreta:
            print(f"¡¡¡Felicidades!!! La palabra secreta era {secreta}. Has acabado con {vidas} vidas",flush = True)

    
    #lo hacemos función para que empiece directamente
    
        automatizar = input("quieres volver a jugar?: Y/N").upper()
        while automatizar == "Y":
            print()
            jugar_ahorcado()
            automatizar = input("quieres volver a jugar?: Y/N").upper()

        if automatizar != "Y":
            print(flush = True)
            print("😊 BYE BYE 😊")    
```
